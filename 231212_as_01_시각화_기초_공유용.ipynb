### 그리드 설정 (grid)


```python
# 보조선 활성화/비활성화 : plt.grid(T/F)
def draw_line(grid=False, style='b-'):
    plt.title('나이와 재산의 상관관계')
    plt.plot(
        range(10, 50, 10), # x축 데이터
        [1, 4, 9, 16], # y축 데이터
        style # 차트의 스타일
    )
    # plt.xlabel('x축 라벨')
    plt.xlabel('나이')
    plt.grid(grid)
    # plt.ylabel('y축 라벨')
    plt.ylabel('재산')
    plt.show()
```

```python
draw_line(grid=True)
draw_line(grid=False)
```

### 여러 개의 선 그어보기

```python
t = np.arange(0., 5., 0.2)
t
```

```python
# 1. 여러 개의 선 그리기 (하나의 plot 사용)
plt.title('여러 개의 라인 플롯')
plt.plot(
    t, t, 'r--',
    t, t **2, 'bs:',
    t, t **3, 'g^-'
)
plt.show()
```

```python
# 2. 하나의 plot이 아니라 여러 개 -> 겹치는.
plt.title('plot 명령 여러번')
plt.plot((np.arange(4) + 1) ** 2)
plt.plot((np.arange(4) - 1) ** 3)
plt.show()
```

```python
# 2. 하나의 plot이 아니라 여러 개 -> 겹치는.
plt.plot((np.arange(4) + 1) ** 2)
plt.show()
plt.plot((np.arange(4) - 1) ** 3)
plt.show()
```

```python
# 여러 개의 플롯이 겹쳤을 경우 범례 표시
X = np.linspace(-np.pi, np.pi, 256)
X
```

```python
# 코사인, 사인 곡선
C, S = np.cos(X), np.sin(X)
C, S
```

```python
plt.title('삼각함수')
plt.plot(X, C)
plt.plot(X, S)
plt.show()
```

```python
plt.title('삼각함수')
plt.plot(X, C, label='cosine')
plt.plot(X, S, label='sine')
plt.legend()
plt.show()
```

#### `subplot`
* plot 명령을 여러 번 실행 => 같은 그래프 내부에 여러 직선 표시
* subplot : 행과 열 단위로 묶어서 여러 데이터를 병행해서 표시

```python
import numpy as np
```

```python
# 위아래 행 배치
plt.plot(np.arange(5) ** 2) # x축 안맞아서 길이 안맞음
plt.title('차트1')
plt.plot(np.arange(10), np.arange(10) ** 3, 'r--')
plt.title('차트2')
plt.show()
```

```python
# 위아래 행 배치
plt.subplot(2, 1, 1)
plt.plot(np.arange(5) ** 2) # x축 안맞아서 길이 안맞음
plt.title('차트1')
plt.subplot(2, 1, 2)
plt.plot(np.arange(10), np.arange(10) ** 3, 'r--')
plt.title('차트2')
plt.tight_layout() # 안겹치게 레이아웃 조정
plt.show()
```

```python
# 양옆 열 배치
# 행, 열, 서브플롯의 위치
# 위아래 행 배치
plt.subplot(2, 1, 1)
plt.plot(np.arange(5) ** 2) # x축 안맞아서 길이 안맞음
plt.title('차트1')
plt.subplot(2, 2, 1)
plt.plot(np.arange(10), np.arange(10) ** 3, 'r--')
plt.title('차트2')
plt.tight_layout() # 안겹치게 레이아웃 조정
plt.show()
```

```python
# 2x2 플롯 배치
plt.subplot(2, 2, 1)
plt.plot(np.arange(5) ** 2) # x축 안맞아서 길이 안맞음
plt.title('차트1')
plt.subplot(2, 2, 2)
plt.plot(np.arange(10), np.arange(10) ** 3, 'r--')
plt.title('차트2')
plt.subplot(2, 2, 3)
plt.plot(np.arange(5) ** 2) # x축 안맞아서 길이 안맞음
plt.title('차트3')
plt.subplot(2, 2, 4)
plt.plot(np.arange(10), np.arange(10) ** 3, 'r--')
plt.title('차트4')
plt.tight_layout() # 안겹치게 레이아웃 조정
plt.show()

```

### 다양한 차트 활용
* `lineplot` 선 차트
* 그외에 여러 가지 차트가 존재

```python
y = [2, 3, 1]
x = np.arange(3)
xlabel = ['가', '나', '다'] # len(x) 길이 같아야함
plt.title('바 차트')
plt.xticks(x, xlabel) # x축의 개별 값에다가 이름 붙여줌
plt.yticks(sorted(y)) # y축에다가 기준이 되는 이름을 붙여주기
plt.bar(x, y)
plt.show()
```

```python
y = [2, 3, 1]
x = np.arange(3)
xlabel = ['가', '나', '다'] # len(x) 길이 같아야함
plt.title('바 차트')
plt.yticks(x, xlabel) # x축의 개별 값에다가 이름 붙여줌
plt.xticks(sorted(y)) # y축에다가 기준이 되는 이름을 붙여주기
plt.barh(x, y)
plt.show()
```

```python
people = ['james', 'john', 'kate', 'alice']
y_pos = np.arange(4)
performance = 3 + 10 * np.random.rand(4)
error = np.random.rand(4)

plt.title('Bar-h Chart')
plt.grid(True)
plt.barh(y_pos, performance, xerr=error, alpha=0.4)
plt.yticks(y_pos, people)
plt.xlabel('x축')
plt.show()
```

#### 파이 차트
* 원형의 파이 모양 그래프

```python
labels = ['호박파이', '사과파이', '정어리파이', '엄마손파이']
# sizes = [15, 30, 45, 10]
sizes = [15, 50, 72, 10] # 비율
# colors = ['yellowgreen', 'gold', 'lightskyblue', 'lightcoral']
colors = ['purple', 'gold', 'lightskyblue', 'lightcoral'] # 색상
explode = (0, 0.1, 0.2, 0.3) # 얼마나 메인에서 떨어질거냐
plt.title("파이 차트")
plt.pie(sizes, explode=explode, labels=labels, colors=colors,
        autopct='%1.1f%%', shadow=True, startangle=90)
plt.axis('equal')
plt.show()
```

### 플롯 종류 선택 (`kind`)

```python
iris : pd.DataFrame = sns.load_dataset('iris') # 붓꽃
titanic : pd.DataFrame = sns.load_dataset('titanic') # 타이타닉호 데이터
```

```python
iris
```

```python
iris.columns
```

```python
# kind = 종류 => 해당 종류 plot 교체
iris['sepal_length'][:20].plot(kind='bar', rot=45, ylim=(4, 6), grid=True, subplots=True)
plt.title('꽃받침의 길이 시각화')
plt.xlabel('Data')
plt.ylabel('꽃받침의 길이')
plt.show()
```

```python
# kind 를 사용할 필요 없이 plot 다음에 직접적으로 해당 차트의 메소드 호출해서 사용
iris[:5].plot.bar(rot=0)
plt.ylim(0, 7)
plt.show()
```

```python
# 파이 차트 : 원형, 비율을 나타낼 때
df2 = titanic['class'].value_counts()
df2
```

```python
df2.plot.pie(autopct="%.2f%%") # 포맷설정
plt.axis('equal') # x축과 y축의 비율 일치
plt.show()
```

```python
# 히스토그램
# - 가로축에 분포(값), 세로축에 빈도(몇개...)를 시각화
iris.plot.hist()
```

```python
# 커널 밀도 그래프
# - 부드럽게 그린 히스토그램
iris.plot.kde()
```
